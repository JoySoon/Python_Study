```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    np.arange(10, 22).reshape(3, 4), # 3행의 4열, 2차원 배열, 행렬
    # columns=['A', 'B', 'C', ] 4열인데 3열에 대한 이름만 주면 에러 발생 (ValueError)
    columns=['A', 'B', 'C', 'D'],
    # index=['a', 'b', 'c', 'd'], # 3행인데 4행에 대한 인덱스 -> 에러 (ValueError)
    index=['a', 'b', 'c'],
)
df
```

```python
#@title .loc 인덱서
#@markdown * 문자열 인덱스 -> 인덱싱, 슬라이싱
#@markdown * .loc[행 인덱스], .loc[행 인덱스, 열 인덱스]
#@markdown * 행 하나만 -> 시리즈
# https://colab.research.google.com/notebooks/forms.ipynb
# row_name = 'a' #@param{"type":"string"}
row_name = 'a' #@param ["a", "b", "c"] {"type":"string"}
df.loc[row_name]
```

```python
# 슬라이싱 : 문자열 -> 끝점이 포함된다
df.loc['a':'b']
df.loc['c'::-1] # 증가폭도 됨
```

```python
df.loc[['c', 'a', 'b']]
```

```python
df.loc[df.A > 15]
```
